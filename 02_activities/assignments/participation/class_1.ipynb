```python
# Import necessary libraries for plotting and data manipulation
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
```

```python
# Generate reproducible data for plotting
np.random.seed(613)

x = np.arange(50)  
y = np.random.randint(0, 100, 50)  

# Create a scatter plot with the generated data
fig, ax = plt.subplots(figsize=(5, 3))  
ax.scatter(x, y)  

plt.show()
```

```python
# Create a bar chart with the same data, adjusting figure size
fig, ax = plt.subplots(figsize=(5, 3))  
ax.bar(x, y)  
```

```python
fig, ax = plt.subplots(figsize=(5, 3))  
ax.barh(x, y)  
```

```python
fig, ax = plt.subplots(figsize=(5, 3))  
ax.plot(x, y)  
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.hist(y)
```

```python
# Create a line chart with title and axis labels
fig, ax = plt.subplots(figsize=(5, 3))  
ax.plot(x, y)  

# Add a title and axis labels
ax.set_title('Total growth over time')  
ax.set_ylabel('Total Growth')          
ax.set_xlabel('Years since start')   
```

```python
# Define custom fonts for title and labels
font1 = {'family': 'sans-serif', 'color': 'blue', 'size': 20}
font2 = {'family': 'monospace', 'color': 'green', 'size': 14}

# Create a line chart with custom fonts
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y)

# Apply custom fonts to title and axis labels
ax.set_title('Total growth over time', fontdict=font1, loc='left')
ax.set_ylabel('Total growth', fontdict=font2)
ax.set_xlabel('Years since start', fontdict=font2)
```

```python
# Create a scatter plot with custom marker and color
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y, 
           marker='*',  # Star-shaped markers
           color='indigo')  # Indigo color for the points
```

```python
# Create a line plot with customized marker, color, linestyle, and linewidth
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y, 
        marker='*',  # Star-shaped markers
        color='indigo',  # Indigo color for the line
        linestyle='--',  # Dashed line style
        linewidth=2)  # Line width set to 2
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y, 
        marker='*',  
        color='indigo',  
        linestyle='--',  
        linewidth=2,  
        markeredgecolor='#FA9359',  # Marker edge color set to a light orange
        markerfacecolor='#000000')  # Marker face color set to black

ax.grid(axis='y', color = 'dodgerblue')
```
