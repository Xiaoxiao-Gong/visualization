```python
import seaborn as sns  
import numpy as np  
import matplotlib.pyplot as plt 
import pandas as pd  
import scipy  
import PIL  # For image processing
import requests  # For making HTTP requests
```

```python
tips = sns.load_dataset("tips")  # Load the 'tips' dataset provided by Seaborn
print(tips)  # Print the dataset to view its contents
```

```python
sns.lineplot(data=tips, x='total_bill', y='tip')  
# Plot 'total_bill' vs 'tip'
```

```python
sns.set_style('whitegrid')  # Set plot style to white grid
sns.lineplot(data=tips, x='total_bill', y='tip')  
```

```python
tipgraph = sns.lineplot(data=tips, 
                        x='total_bill', 
                        y='tip')  # Plot 'total_bill' vs 'tip'
tipgraph.set(title='Tips vs. Total Bill', 
             xlabel='Total Bill ($)', 
             ylabel='Tip Amount ($)')  
# Add title and axis labels
```

```python
fig = plt.subplots(figsize=(10, 3))  # Set figure size

tipgraph = sns.lineplot(data=tips, 
                        x='total_bill', 
                        y='tip',  # Plot 'total_bill' vs 'tip'
                        color='hotpink', 
                        linestyle='--', 
                        linewidth=3,  # Customize line style and color
                        marker='o', 
                        markerfacecolor='indigo'  # Customize marker style and color
)
tipgraph.set(title='Tips vs. Total Bill', 
             xlabel='Total Bill ($)', 
             ylabel='Tip Amount ($)'
)  # Add title and labels
```

```python
tipgraph = sns.scatterplot(data=tips, 
                           x='total_bill', 
                           y='tip',  # Plot 'total_bill' vs 'tip'
                           style='time', 
                           hue='day',  # Differentiate points by 'time' and 'day'
                           palette=['purple', 'hotpink', 'deepskyblue', 'yellowgreen']  # Custom colors
)
tipgraph.set(title='Tip vs. Total Bill', 
             xlabel='Total Bill ($)', 
             ylabel='Tip Amount ($)'
)  # Add title and labels
```

```python
sns.pairplot(data=tips, hue='day')  # Pairwise plots differentiated by 'day'
```

```python
daysplot = sns.relplot(
    data=tips, x='total_bill', y='tip',  # Plot 'total_bill' vs 'tip'
    hue='sex', col='day',  # Differentiate by 'sex', create subplots by 'day'
    kind='scatter', col_wrap=2  # Scatter plot, wrap columns into 2 per row
)
daysplot  # Display the plot
```

```python
import plotly.graph_objects as go  # Import Plotly for creating plots

x1 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])  # Character names
y1 = np.array([110, 180, 240, 99, 220])  # Corresponding values
```

```python
graph = go.Figure()  # Create a new figure
graph.add_trace(go.Bar(x=x1, y=y1))  # Add a bar chart with x1 and y1 data

graph.update_layout(
    title="Pirate Scores",  
    xaxis_title="Pirates",  
    yaxis_title="Score"  
)
```

```python
graph = go.Figure()  # Create a new figure

graph.add_trace(go.Scatter(
    x=x1, y=y1,  # Set x and y data
    mode='markers',  # Use markers for the plot
    marker=dict(
        size=15, color='hotpink',  # Customize marker size and color
        opacity=1,  # Set marker opacity
        line=dict(width=5, color='purple')  # Customize marker outline
    )
))

graph.update_layout(
    title="Interactive Pirate Plot",  
    xaxis_title="Pirates",  
    yaxis_title="Scores",  
    width=500, height=500  # Set figure dimensions
)
```

```python
from wordcloud import WordCloud  # Import WordCloud for generating word clouds

df = pd.read_csv(
    "https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv",
    on_bad_lines='skip'  # Skip lines with errors while reading the CSV file
)

df
```

```python
text = " ".join(each for each in df.quote)  # Combine all quotes into a single string

wordcloud = WordCloud(
    background_color='white', colormap='inferno'  # Set background and color theme
).generate(text)  # Generate word cloud from text

fig, ax = plt.subplots(figsize=(7, 3))  # Set figure size
ax.imshow(wordcloud, interpolation='bilinear')  # Display the word cloud
ax.axis("off")  # Turn off axis display
```

```python
from matplotlib_venn import venn2  # Import Venn diagram function

A = set(['apple', 'banana', 'watermelon'])  # Define set A
B = set(['pumpkin', 'blueberry', 'apple', 'key lime'])  # Define set B

diagram = venn2(
    [A, B],  # Sets to compare
    set_labels=('Fruits', 'Pies'),  # Labels for the sets
    set_colors=('blue', 'red'),  # Colors for the sets
    alpha=0.5  # Transparency of the sets
)

diagram.get_label_by_id("10").set_text("\n".join(A - B))  
# Set label for elements unique to set A

diagram.get_label_by_id("11").set_text("\n".join(A & B))  
# Set label for elements common to both sets A and B

diagram.get_label_by_id("01").set_text("\n".join(B - A))  
# Set label for elements unique to set B
```
